```python
df1=pd.DataFrame({"Employee_ID":[104,102,103,104],"name":["alice","bob","carol","dan"],"Department":["Hr","it","Marketing","it"]})
df2=pd.DataFrame({"Employee_ID":[101,102,105,106],"Salary":[50000,70000,45000,40000]})

merged=pd.merge(df1,df2,on="Employee_ID",how="outer")
##merged=pd.concat([df1,df2],ignore_index=True)
merged


```

```python
dep_it=merged[merged['Department']=='it']
avg_it=dep_it['Salary'].mean()
sum_it=dep_it['Salary'].sum()

print(avg_it)

print(sum_it)
```

```python
merged[merged['Salary']>60000]
merged
```

```python
def hike(row):      # 1st row in the DataFrame
    dept = row['Department']   #department col in 1st row
    if isinstance(dept, str):  # Only proceed if it's a string
        dept = dept.lower()       # convert to lower case
        if dept == 'it':
            return row['Salary'] * 0.10
        elif dept in ['hr', 'marketing']:
            return row['Salary'] * 0.05
    return 0  # No bonus for other departments or missing department
    
merged['Bonus'] = merged.apply(hike, axis=1)
```
